```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns

# Generating random data
np.random.seed(42)

# Random data for three companies over 12 months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
companies = ['Apple', 'Samsung', 'Huawei']

data = {
    'Month': months,
    'Apple Sales': np.random.randint(1000, 5000, 12),
    'Samsung Sales': np.random.randint(1000, 5000, 12),
    'Huawei Sales': np.random.randint(1000, 5000, 12),
    'Apple Market Share': np.random.uniform(0.2, 0.4, 12),
    'Samsung Market Share': np.random.uniform(0.2, 0.4, 12),
    'Huawei Market Share': np.random.uniform(0.1, 0.3, 12),
    'Apple Customer Satisfaction': np.random.uniform(70, 90, 12),
    'Samsung Customer Satisfaction': np.random.uniform(60, 85, 12),
    'Huawei Customer Satisfaction': np.random.uniform(50, 80, 12),
}

df = pd.DataFrame(data)

# Scaling data for radar plot
scaler = MinMaxScaler()

apple_data = scaler.fit_transform(df[['Apple Sales', 'Apple Market Share', 'Apple Customer Satisfaction']].values)
samsung_data = scaler.fit_transform(df[['Samsung Sales', 'Samsung Market Share', 'Samsung Customer Satisfaction']].values)
huawei_data = scaler.fit_transform(df[['Huawei Sales', 'Huawei Market Share', 'Huawei Customer Satisfaction']].values)

```

```python
# Plotting each chart separately

# 1. Line Plot: Apple Sales
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Apple Sales'], label='Apple Sales')
plt.title('Apple Sales (Line Plot)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()

```

```python
# 2. Bar Chart: Samsung Sales
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['Samsung Sales'], label='Samsung Sales', color='orange')
plt.title('Samsung Sales (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()

```

```python
# 3. Violin Plot: Huawei Customer Satisfaction
plt.figure(figsize=(10, 6))
plt.violinplot(df['Huawei Customer Satisfaction'])
plt.title('Huawei Customer Satisfaction (Violin Plot)')
plt.ylabel('Satisfaction')
plt.show()

```

```python
# 4. Pie Chart: Apple Market Share Distribution
plt.figure(figsize=(10, 6))
market_share_apple = df['Apple Market Share'].iloc[-1]
market_share_samsung = df['Samsung Market Share'].iloc[-1]
market_share_huawei = df['Huawei Market Share'].iloc[-1]
plt.pie([market_share_apple, market_share_samsung, market_share_huawei], labels=['Apple', 'Samsung', 'Huawei'], autopct='%1.1f%%')
plt.title('Apple Market Share (Pie Chart)')
plt.show()

```

```python
# 5. Boxplot: Apple Sales Distribution
plt.figure(figsize=(10, 6))
plt.boxplot(df['Apple Sales'])
plt.title('Apple Sales (Box Plot)')
plt.ylabel('Sales')
plt.show()

```

```python
# 6. Radar Chart: Apple, Samsung, Huawei Comparison
plt.figure(figsize=(10, 6))
categories = ['Sales', 'Market Share', 'Customer Satisfaction']
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()

apple_data_mean = np.mean(apple_data, axis=0)
samsung_data_mean = np.mean(samsung_data, axis=0)
huawei_data_mean = np.mean(huawei_data, axis=0)

plt.polar(angles, apple_data_mean, label='Apple', color='blue')
plt.fill(angles, apple_data_mean, color='blue', alpha=0.1)
plt.polar(angles, samsung_data_mean, label='Samsung', color='orange')
plt.fill(angles, samsung_data_mean, color='orange', alpha=0.1)
plt.polar(angles, huawei_data_mean, label='Huawei', color='green')
plt.fill(angles, huawei_data_mean, color='green', alpha=0.1)
plt.title('Brand Comparison (Radar Chart)')
plt.yticks([])
plt.xticks(angles, categories)
plt.legend()
plt.show()

```

```python
# 7. Scatter Plot: Samsung Sales vs. Customer Satisfaction
plt.figure(figsize=(10, 6))
plt.scatter(df['Samsung Sales'], df['Samsung Customer Satisfaction'], label='Samsung', color='orange')
plt.title('Samsung Sales vs. Customer Satisfaction (Scatter Plot)')
plt.xlabel('Sales')
plt.ylabel('Customer Satisfaction')
plt.show()

```

```python
# 8. Histogram: Apple Customer Satisfaction Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Apple Customer Satisfaction'], bins=10, alpha=0.7, color='blue')
plt.title('Apple Customer Satisfaction (Histogram)')
plt.xlabel('Satisfaction')
plt.ylabel('Frequency')
plt.show()

```

```python
# 9. Heatmap: Correlation between Sales and Market Share
plt.figure(figsize=(10, 6))
corr_data = df[['Apple Sales', 'Samsung Sales', 'Huawei Sales', 'Apple Market Share', 'Samsung Market Share', 'Huawei Market Share']].corr()
sns.heatmap(corr_data, annot=True, cmap='coolwarm')
plt.title('Sales & Market Share Correlation (Heatmap)')
plt.show()
```
