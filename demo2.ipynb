```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

n = 100 

# Simulate financial data for 100 observations
store_investment = np.random.randint(800000, 1200000, n)  # Investment per store
revenue = np.random.randint(1200000, 1800000, n)  # Revenue per store in the first year
capital_raised = np.random.randint(200000, 500000, n)  # Capital raised for expansion
profit_margin = np.random.uniform(0.1, 0.3, n)  # Profit margin per store

# Simulate market data for competition and demand
competition_score = np.random.uniform(0.2, 1.0, n)  # 1 is very competitive, 0 is low competition
consumer_awareness = np.random.uniform(0.5, 1.0, n)  # Awareness of organic products
organic_demand = np.random.uniform(0.5, 1.0, n)  # Demand for organic food
price_sensitivity = np.random.uniform(0.3, 0.7, n)  # Price sensitivity of consumers
store_cost = store_investment + np.random.randint(100000, 500000, n)  # Total store cost

# Create a DataFrame with all the data
data = pd.DataFrame({
    'store_investment': store_investment,
    'revenue': revenue,
    'capital_raised': capital_raised,
    'profit_margin': profit_margin,
    'competition_score': competition_score,
    'consumer_awareness': consumer_awareness,
    'organic_demand': organic_demand,
    'price_sensitivity': price_sensitivity,
    'store_cost': store_cost
})

# Plotting the results
fig, axs = plt.subplots(4, 2, figsize=(14, 10))

# 1. Store Investment vs. Revenue
axs[0, 0].scatter(data['store_investment'], data['revenue'], color='blue')
axs[0, 0].set_title('Store Investment vs. Revenue')
axs[0, 0].set_xlabel('Store Investment ()')
axs[0, 0].set_ylabel('Revenue ()')

# 2. Profit Margin Distribution
axs[0, 1].hist(data['profit_margin'], bins=15, color='green', edgecolor='black')
axs[0, 1].set_title('Profit Margin Distribution')
axs[0, 1].set_xlabel('Profit Margin')
axs[0, 1].set_ylabel('Frequency')

# 3. Capital Raised vs. Store Investment
axs[1, 0].scatter(data['capital_raised'], data['store_investment'], color='red')
axs[1, 0].set_title('Capital Raised vs. Store Investment')
axs[1, 0].set_xlabel('Capital Raised ()')
axs[1, 0].set_ylabel('Store Investment ()')

# 4. Consumer Awareness vs. Organic Demand
axs[1, 1].scatter(data['consumer_awareness'], data['organic_demand'], color='purple')
axs[1, 1].set_title('Consumer Awareness vs. Organic Demand')
axs[1, 1].set_xlabel('Consumer Awareness')
axs[1, 1].set_ylabel('Organic Demand')

# 5. Price Sensitivity vs. Competition Score
axs[2, 0].scatter(data['price_sensitivity'], data['competition_score'], color='orange')
axs[2, 0].set_title('Price Sensitivity vs. Competition Score')
axs[2, 0].set_xlabel('Price Sensitivity')
axs[2, 0].set_ylabel('Competition Score')

# 6. Store Cost vs. Revenue
axs[2, 1].scatter(data['store_cost'], data['revenue'], color='brown')
axs[2, 1].set_title('Store Cost vs. Revenue')
axs[2, 1].set_xlabel('Store Cost ()')
axs[2, 1].set_ylabel('Revenue ()')

# 7. Market Demand vs. Profit Margin
axs[3, 0].scatter(data['organic_demand'], data['profit_margin'], color='pink')
axs[3, 0].set_title('Market Demand vs. Profit Margin')
axs[3, 0].set_xlabel('Market Demand')
axs[3, 0].set_ylabel('Profit Margin')

# 8. Distribution of Store Costs
axs[3, 1].hist(data['store_cost'], bins=15, color='blue', edgecolor='black')
axs[3, 1].set_title('Distribution of Store Costs')
axs[3, 1].set_xlabel('Store Costs ()')
axs[3, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
